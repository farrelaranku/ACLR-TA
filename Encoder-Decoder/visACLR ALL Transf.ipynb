```python

data_dict = {
    '96': w96,
    '192': w192,
    '336': w336,
    '720': w720
}

fig, axs = plt.subplots(2, 2, figsize=(12, 8))
fig.suptitle('Learning Rate per Iteration', fontsize=14)

for i, (key, df) in enumerate(data_dict.items()):
    row, col = divmod(i, 2)
    ax = axs[row][col]
    half_len = len(df) // 2
    ax.plot(df.iloc[:half_len, 2], label=f'Pred Len {key}')
    ax.set_title(f'Prediction Length: {key}')
    ax.set_xlabel('Iteration')
    ax.set_ylabel('Learning Rate')
    ax.legend()
    ax.grid(True)

plt.tight_layout(rect=[0, 0, 1, 0.96])  # Biar title utama ga ketimpa
plt.show()
```
